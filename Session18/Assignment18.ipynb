```python
'''
Q1. (Data Loading & Initial Analysis)

Load the Ford Car Dataset using pandas.
Display the first 10 rows and last 5 rows.
Check the shape (number of rows and columns) and data types of all columns.
Write observations about the dataset structure in comments.
'''

import pandas as pd

# Load the dataset
df = pd.read_csv("ford_car_dataset.csv")

# Display first 10 rows
print("First 10 Rows:")
print(df.head(10))

# Display last 5 rows
print("\nLast 5 Rows:")
print(df.tail())

# Display shape
print("\nShape of Dataset:")
print(df.shape)

# Display data types
print("\nData Types:")
print(df.dtypes)

# Observations:
# 1. The dataset contains information about Ford cars and their features.
# 2. It consists of both numerical and categorical columns.
# 3. The shape indicates the total number of records and features.
# 4. Columns such as model and transmission are categorical, while year, price,
#    mileage, mpg, engineSize, and tax are numerical.
# 5. The dataset is suitable for data analysis and regression-based machine learning.
```

```python
'''
Q2. (Missing & Duplicate Values)

Perform a detailed check on the dataset:
Find the total number of missing values in each column.
Identify and remove duplicate rows if any.
Write a summary of your findings and how you handled them in comments.
'''

# Check missing values
print("Missing Values in Each Column:")
print(df.isnull().sum())

# Check duplicate rows
duplicate_count = df.duplicated().sum()
print("\nNumber of Duplicate Rows:", duplicate_count)

# Remove duplicate rows (if any)
df = df.drop_duplicates()

# Verify duplicates after removal
print("\nDuplicate Rows After Removal:")
print(df.duplicated().sum())

# Observations:
# 1. Checked all columns for missing values using isnull().sum().
# 2. Counted duplicate records using duplicated().sum().
# 3. Removed duplicate rows using drop_duplicates().
# 4. Verified that no duplicate rows remain in the dataset.
# 5. Removing duplicates improves data quality for analysis and machine learning.
```

```python
'''
Q3. (Statistical Summary)

Generate the statistical summary of all numeric columns using describe().
Identify minimum, maximum, mean, and median values for key columns like
price, mileage, and year.
'''

# Display statistical summary
print("Statistical Summary:")
print(df.describe())

# Minimum, Maximum, Mean, and Median for selected columns
columns = ["price", "mileage", "year"]

for col in columns:
    print(f"\n{col.capitalize()} Statistics")
    print("Minimum :", df[col].min())
    print("Maximum :", df[col].max())
    print("Mean    :", df[col].mean())
    print("Median  :", df[col].median())

# Observations:
# 1. The statistical summary provides an overview of all numeric columns.
# 2. Mean and median indicate the central tendency of the data.
# 3. Minimum and maximum values show the range of each feature.
# 4. These statistics help in understanding the distribution before further analysis.
```

```python
'''
Q4. (Histogram of Numeric Features)

Plot histograms for all important numeric columns (price, mileage, year,
engineSize, mpg).
Analyze the distribution for each feature and note your findings in comments.
'''
# Import Library
import matplotlib.pyplot as plt

# List of numeric columns
numeric_columns = ["price", "mileage", "year", "engineSize", "mpg"]

# Plot histograms
for col in numeric_columns:
    plt.figure(figsize=(7, 4))
    plt.hist(df[col], bins=20, edgecolor="black", color="skyblue")
    plt.title(f"Histogram of {col}")
    plt.xlabel(col)
    plt.ylabel("Frequency")
    plt.grid(axis="y", linestyle="--", alpha=0.6)
    plt.show()

# Observations:
# 1. Price values are spread over a wide range, with most cars in the mid-price range.
# 2. Mileage is right-skewed, indicating that many cars have lower mileage while fewer have very high mileage.
# 3. The year distribution shows that most cars are from recent manufacturing years.
# 4. Engine size is concentrated around common values such as 1.0L, 1.5L, and 2.0L.
# 5. MPG values are mostly clustered within a specific range, with a few extreme values.
```

```python
'''
Q5. (Count Plots of Categorical Features)

Create count plots for all categorical columns (fuelType, transmission, model, etc.)
using seaborn.
Analyze which categories are most common.

Write insights about market trends based on these plots in comments.
'''
# Import Library
import seaborn as sns

# Select categorical columns
categorical_columns = ["model", "transmission", "fuelType"]

# Plot count plots
for col in categorical_columns:
    plt.figure(figsize=(10, 5))

    sns.countplot(
        data=df,
        x=col,
        hue=col,
        order=df[col].value_counts().index,
        palette="Set2",
        legend=False
    )

    plt.title(f"Count Plot of {col}")
    plt.xlabel(col)
    plt.ylabel("Count")
    plt.xticks(rotation=45)
    plt.show()

# Observations:
# 1. Count plots show the frequency of each category in the dataset.
# 2. Some Ford models appear much more frequently than others.
# 3. Manual and Automatic are the most common transmission types.
# 4. Petrol and Diesel vehicles dominate the dataset, while other fuel types are less common.
# 5. These trends indicate higher market availability and customer preference for these categories.
```

```python
'''
Q6. (Correlation Heatmap)

Create a heatmap of the correlation matrix for numeric features.
Use seaborn with proper annotations.
Identify which features are highly correlated with the target variable (price).
Write your observations in comments.
'''

# Select only numeric columns
numeric_df = df.select_dtypes(include=["number"])

# Create correlation matrix
corr_matrix = numeric_df.corr()

# Plot heatmap
plt.figure(figsize=(10, 8))

sns.heatmap(
    corr_matrix,
    annot=True,
    cmap="coolwarm",
    fmt=".2f",
    linewidths=0.5
)

plt.title("Correlation Heatmap")
plt.show()

# Display correlation with price
print("Correlation with Price:")
print(corr_matrix["price"].sort_values(ascending=False))

# Observations:
# 1. The heatmap shows the correlation between all numeric features.
# 2. Features with higher positive correlation are more closely related to price.
# 3. Mileage generally has a negative correlation with price, indicating that
#    cars with higher mileage tend to have lower prices.
# 4. Year and engineSize usually show a positive correlation with price,
#    meaning newer cars and larger engines often have higher prices.
# 5. Correlation analysis helps identify important features for regression models.
```

```python
'''
Q7. (Feature Identification)

Clearly identify and list:
Independent Features (Input Features)
Dependent Feature (Output/Target Feature)
Justify your choice with reasoning.
'''

# Independent (Input) Features
X = df.drop("price", axis=1)

# Dependent (Target) Feature
y = df["price"]

print("Independent Features (Input Features):")
print(X.columns.tolist())

print("\nDependent Feature (Target Feature):")
print(y.name)

# Observations:
# 1. The independent features are all columns except 'price'.
# 2. The dependent feature is 'price' because it is the value we want to predict.
# 3. Features such as year, mileage, transmission, fuelType, mpg, and engineSize
#    influence the selling price of a Ford car.
# 4. Therefore, 'price' is selected as the target variable for building a
#    regression model.
```

```python
'''
Q8. (Encoding Categorical Variables)

Convert all categorical columns into numeric form using appropriate encoding
techniques:
Use One Hot Encoding.
Show before and after transformation for at least 2 columns.
'''

# Display original categorical columns
print("Before One Hot Encoding:")
print(df[["model", "transmission"]].head())

# Apply One Hot Encoding
encoded_df = pd.get_dummies(
    df,
    columns=["model", "transmission", "fuelType"],
    drop_first=True
)

# Display encoded columns
print("\nAfter One Hot Encoding (Sample):")
print(encoded_df.filter(regex="model_|transmission_").head())

print("\nShape Before Encoding:", df.shape)
print("Shape After Encoding:", encoded_df.shape)

# Observations:
# 1. One Hot Encoding converts categorical values into binary (0/1) columns.
# 2. The original categorical columns are replaced with new encoded columns.
# 3. The number of columns increases after encoding because each category
#    becomes a separate feature.
# 4. This transformation makes the dataset suitable for machine learning models.
```

```python
'''
Q9. (Feature Scaling)

Apply Standard Scaling (using StandardScaler from sklearn) on the numeric
independent features.
Show the first 5 rows of scaled data.
'''

from sklearn.preprocessing import StandardScaler

# Select numeric independent features
numeric_features = ["year", "mileage", "tax", "mpg", "engineSize"]

# Create StandardScaler object
scaler = StandardScaler()

# Apply feature scaling
scaled_data = scaler.fit_transform(df[numeric_features])

# Convert scaled data into a DataFrame
scaled_df = pd.DataFrame(scaled_data, columns=numeric_features)

# Display first 5 rows
print("First 5 Rows of Scaled Data:")
print(scaled_df.head())

# Observations:
# 1. Standard Scaling transforms the data so that each feature has
#    a mean close to 0 and a standard deviation close to 1.
# 2. Scaling helps improve the performance of many machine learning algorithms.
# 3. Only numeric independent features are scaled, while categorical features
#    are handled separately using encoding.
```

```python
'''
Q10. (Mini Project - Complete Preprocessing Pipeline)

Create a complete preprocessing pipeline for the Ford Car Dataset:
1. Load and clean the data (handle missing & duplicate values).
2. Perform EDA (histograms + count plots + heatmap).
3. Identify input and output features.
4. Encode categorical variables.
5. Scale the numeric features.
'''

from sklearn.preprocessing import StandardScaler

# Step 1: Data Cleaning
print("Missing Values:")
print(df.isnull().sum())

df = df.drop_duplicates()
print("\nDuplicate Rows After Removal:", df.duplicated().sum())

# Step 2: Exploratory Data Analysis (EDA)

# Histogram
df[["price", "mileage", "year", "engineSize", "mpg"]].hist(
    figsize=(12, 8),
    bins=20,
    edgecolor="black"
)
plt.suptitle("Histograms of Numeric Features")
plt.show()

# Count Plots
categorical_columns = ["model", "transmission", "fuelType"]

for col in categorical_columns:
    plt.figure(figsize=(8, 4))
    sns.countplot(
        data=df,
        x=col,
        hue=col,
        palette="Set2",
        legend=False,
        order=df[col].value_counts().index
    )
    plt.title(f"Count Plot of {col}")
    plt.xticks(rotation=45)
    plt.show()

# Correlation Heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(
    df.select_dtypes(include="number").corr(),
    annot=True,
    cmap="coolwarm",
    fmt=".2f"
)
plt.title("Correlation Heatmap")
plt.show()

# Step 3: Identify Features
X = df.drop("price", axis=1)
y = df["price"]

print("\nInput Features:")
print(X.columns.tolist())

print("\nOutput Feature:")
print(y.name)

# Step 4: One Hot Encoding
X_encoded = pd.get_dummies(
    X,
    columns=["model", "transmission", "fuelType"],
    drop_first=True
)

# Step 5: Feature Scaling
numeric_columns = ["year", "mileage", "tax", "mpg", "engineSize"]

scaler = StandardScaler()

X_encoded[numeric_columns] = scaler.fit_transform(
    X_encoded[numeric_columns]
)

print("\nFirst 5 Rows of Preprocessed Data:")
print(X_encoded.head())

# Summary:
# 1. Loaded and cleaned the dataset by checking missing values and removing duplicates.
# 2. Performed EDA using histograms, count plots, and a correlation heatmap.
# 3. Selected 'price' as the target variable and the remaining columns as input features.
# 4. Applied One Hot Encoding to convert categorical features into numeric form.
# 5. Applied Standard Scaling to the numeric features.
# 6. The dataset is now ready for training a machine learning regression model.
```
