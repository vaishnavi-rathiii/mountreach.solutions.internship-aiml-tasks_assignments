## Special Task – Spotify Tracks Dataset Analysis

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Optional: Better plot style
plt.style.use("ggplot")
sns.set_theme(style="whitegrid")
```

```python
plt.figure(figsize=(12,6))

top_genres = df["track_genre"].value_counts().head(10)

top_genres.plot(kind="bar", color="teal")

plt.title("Top 10 Music Genres")
plt.xlabel("Genre")
plt.ylabel("Number of Songs")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(8,5))

plt.hist(df["popularity"], bins=20, edgecolor="black")

plt.title("Distribution of Song Popularity")
plt.xlabel("Popularity")
plt.ylabel("Frequency")

plt.show()
```

```python
plt.figure(figsize=(6,6))

df["explicit"].value_counts().plot(
    kind="pie",
    autopct="%1.1f%%",
    startangle=90
)

plt.title("Explicit vs Non-Explicit Songs")
plt.ylabel("")
plt.show()
```

```python
plt.figure(figsize=(6,5))
sns.boxplot(y=df["danceability"])
plt.title("Danceability Distribution")
plt.show()
```

```python
plt.figure(figsize=(8,6))

sns.scatterplot(
    x="energy",
    y="popularity",
    data=df,
    alpha=0.5
)

plt.title("Energy vs Popularity")

plt.show()
```

```python
plt.figure(figsize=(12,8))

numeric_df = df.select_dtypes(include="number")

sns.heatmap(
    numeric_df.corr(),
    annot=True,
    cmap="coolwarm",
    fmt=".2f"
)

plt.title("Correlation Heatmap")

plt.show()
```

```python
plt.figure(figsize=(10,6))

top_artists = df["artists"].value_counts().head(10)
top_artists.plot(kind="barh", color="orange")

plt.title("Top 10 Artists with Most Songs")
plt.xlabel("Number of Songs")
plt.show()
```

### Dataset Summary

The **Spotify Tracks Dataset** contains information about Spotify songs, including their audio features, artist details, genre, and popularity. It helps analyze music trends and understand how different song characteristics are related.

### Key Insights

* The dataset contains songs from multiple artists and genres.
* Most songs have medium popularity.
* Danceability and energy are generally high for many songs.
* Some audio features are positively correlated, while others show weak relationships.
* The dataset is suitable for music analysis and machine learning applications.
