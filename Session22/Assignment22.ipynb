```python
'''
Q8. (Loading & Testing Saved Model)

Load the saved model and preprocessing objects.
Create sample input data (one patient record).
Preprocess the input (encoding + scaling).
Make a prediction using the loaded model and print the result.
'''

import pandas as pd
import joblib

# Load saved objects
model = joblib.load("heart_model.pkl")
scaler = joblib.load("scaler.pkl")
encoded_columns = joblib.load("columns.pkl")

# Create sample patient data
sample = pd.DataFrame({
    "Age": [45],
    "Sex": ["M"],
    "ChestPainType": ["ATA"],
    "RestingBP": [130],
    "Cholesterol": [230],
    "FastingBS": [0],
    "RestingECG": ["Normal"],
    "MaxHR": [150],
    "ExerciseAngina": ["N"],
    "Oldpeak": [1.2],
    "ST_Slope": ["Up"]
})

# Perform One-Hot Encoding
sample = pd.get_dummies(sample)

# Align columns with training data
sample = sample.reindex(columns=encoded_columns, fill_value=0)

# Apply Feature Scaling
sample_scaled = scaler.transform(sample)

# Convert scaled data to DataFrame
sample_scaled = pd.DataFrame(sample_scaled, columns=encoded_columns)

# Make Prediction
prediction = model.predict(sample_scaled)

# Display Result
if prediction[0] == 1:
    print("Prediction: Heart Disease Detected")
else:
    print("Prediction: No Heart Disease")
```
