```python
'''
Q2. (Dataset Selection & Initial Exploration)

Load the dataset and show first 10 rows.
Write 5-6 lines explaining why you chose this dataset and what continuous
value you want to predict.
'''

import pandas as pd

# Load the dataset
df = pd.read_csv("House_Price_Prediction.csv")

# Display first 10 rows
print("First 10 Rows:")
print(df.head(10))

"""
Why I Chose This Dataset:

The House Price Prediction dataset contains a continuous target variable, 'price',
which makes it suitable for regression analysis. It includes important features
such as bedrooms, bathrooms, living area, and location that influence house
prices. This dataset provides a good opportunity to perform data preprocessing,
exploratory data analysis, and visualization. The objective is to predict the
selling price of a house based on its characteristics using machine learning.
"""
```

```python
'''
Q3. (Missing & Duplicate Values Analysis)

Check for missing values in each column and calculate their percentage.
Handle missing values appropriately (drop or fill).
Check and remove duplicate rows if any.
Write your observations and steps taken.
'''

# Check missing values
missing_values = df.isnull().sum()

# Calculate missing value percentage
missing_percentage = (missing_values / len(df)) * 100

print("Missing Values:")
print(missing_values)

print("\nMissing Value Percentage:")
print(missing_percentage)

# Fill missing values
for column in df.columns:
    if df[column].dtype == "object":
        df[column] = df[column].fillna(df[column].mode()[0])
    else:
        df[column] = df[column].fillna(df[column].median())

# Check duplicate rows
duplicate_count = df.duplicated().sum()

print("\nNumber of Duplicate Rows:", duplicate_count)

# Remove duplicate rows
df = df.drop_duplicates()

# Verify duplicates after removal
print("Duplicate Rows After Removal:", df.duplicated().sum())

"""
Observations:

1. Checked all columns for missing values and calculated their percentage.
2. Filled numerical missing values using the median.
3. Filled categorical missing values using the mode.
4. Identified and removed duplicate records.
5. The dataset was cleaned and is ready for further analysis and preprocessing.
"""
```

```python
'''
Q4. (Statistical Summary)

Use describe() to generate statistical summary of numeric columns.
Identify min, max, mean, and median of the target variable.
Write at least 4 meaningful observations about the data distribution.
'''

# Display statistical summary
print("Statistical Summary:")
print(df.describe())

# Target variable statistics
print("\nPrice Statistics")
print("Minimum :", df["price"].min())
print("Maximum :", df["price"].max())
print("Mean    :", df["price"].mean())
print("Median  :", df["price"].median())

"""
Observations:

1. The house prices have a wide range, indicating high variation in property values.
2. The mean price is higher than the median, suggesting the presence of high-priced houses (right-skewed distribution).
3. The minimum price is 0, which may represent an invalid or missing value and should be investigated.
4. The large difference between the minimum and maximum values indicates the presence of potential outliers in the dataset.
"""
```

```python
'''
Q6. (Count Plots of Categorical Features)

Identify categorical columns in your dataset.
Create count plots for all categorical features using seaborn.
Analyze the distribution and write business/domain-related insights.
'''

import matplotlib.pyplot as plt
import seaborn as sns

# Identify categorical columns
categorical_columns = df.select_dtypes(include="object").columns

print("Categorical Columns:")
print(categorical_columns.tolist())

# Plot count plots
for col in categorical_columns:
    plt.figure(figsize=(12, 5))

    sns.countplot(
        y=col,
        data=df,
        order=df[col].value_counts().index[:10],   # Top 10 categories
        hue=col,
        palette="viridis",
        legend=False
    )

    plt.title(f"Count Plot of {col}")
    plt.xlabel("Count")
    plt.ylabel(col)
    plt.tight_layout()
    plt.show()

"""
Business/Domain Insights:

• Street
- Some streets have significantly more house listings than others.
- Areas with more listings may indicate higher residential demand.

• City
- Certain cities contain a larger number of properties.
- Cities with more listings may represent active real estate markets.

• Statezip
- Property distribution differs across postal codes.
- Some locations are more popular, which may influence house prices.

• Country
- Almost all properties belong to the same country.
- This feature provides little variation and may have limited impact on prediction.
"""
```

```python
'''
Q7. (Correlation Heatmap)

Create a correlation heatmap for all numeric features.
Highlight which features are strongly correlated with the target variable.
Write your observations regarding feature relationships.
'''

import matplotlib.pyplot as plt
import seaborn as sns

# Select numeric columns
numeric_df = df.select_dtypes(include=["int64", "float64"])

# Correlation matrix
corr_matrix = numeric_df.corr()

# Plot heatmap
plt.figure(figsize=(12, 8))

sns.heatmap(
    corr_matrix,
    annot=True,
    cmap="coolwarm",
    fmt=".2f"
)

plt.title("Correlation Heatmap")
plt.show()

"""
Observations:

1. The target variable 'price' shows a strong positive correlation with
   sqft_living and bathrooms.

2. Bedrooms have a moderate positive correlation with price.

3. Sqft_above is highly correlated with sqft_living because it forms a major
   part of the total living area.

4. Yr_built has a positive correlation with price, indicating that newer
   houses generally tend to have higher prices.

5. Features such as waterfront and view may also show positive correlations,
   suggesting that houses with better views or waterfront locations are often
   more expensive.
"""
```

```python
'''
Q9. (Encoding Categorical Variables)

Convert all categorical columns to numeric form using suitable encoding
techniques (One-Hot Encoding).
Show before and after transformation for at least two columns.
Explain your encoding choices.
'''

# Display categorical columns
categorical_columns = df.select_dtypes(include="object").columns

print("Categorical Columns:")
print(categorical_columns.tolist())

# Before Encoding (Two Columns)
print("\nBefore Encoding:")
print(df[["city", "country"]].head())

# Apply One-Hot Encoding
df_encoded = pd.get_dummies(df, columns=categorical_columns, drop_first=True)

# After Encoding (Sample Columns)
print("\nAfter Encoding:")
encoded_columns = [col for col in df_encoded.columns if "city_" in col or "country_" in col]
print(df_encoded[encoded_columns[:10]].head())

"""
Explanation:

1. One-Hot Encoding converts categorical values into binary (0 and 1) columns.
2. It prevents the model from assuming any numerical order between categories.
3. The 'drop_first=True' parameter removes one category from each feature to
   avoid multicollinearity (dummy variable trap).
4. The encoded dataset is now suitable for machine learning algorithms.
"""
```
